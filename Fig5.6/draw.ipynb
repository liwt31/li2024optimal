```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
plt.style.use(['science'])
import matplotlib as mpl
mpl.rcParams['xtick.labelsize'] = "medium"
mpl.rcParams['ytick.labelsize'] = "medium"
mpl.rcParams['legend.fontsize'] = "large"
mpl.rcParams['legend.frameon'] = True
mpl.rcParams['legend.framealpha'] = 1
mpl.rcParams['lines.linewidth'] = 1.5
```

```python
fig = plt.figure(figsize=(10, 3))
axis = fig.subplots(1, 3)

plt.sca(axis[0])

n_list = ["004", "008", "016", "032", "064", "128", "256"]
n_array = np.array(list(map(float, n_list)))

t_list = []
std_list = []
for n in n_list:
    with open(f"vsN/ps1_linear_{n}.log") as fin:
        lines = fin.readlines()
    t_list.append(float(lines[-1].split()[-2][1:-1]))
    std_list.append(float(lines[-1].split()[-1][:-1]))
plt.plot(n_array, t_list, marker="o", label="MPS")

t_list = []
std_list = []
for n in n_list:
    with open(f"vsN/ps1_binary_{n}.log") as fin:
        lines = fin.readlines()
    t_list.append(float(lines[-1].split()[-2][1:-1]))
    std_list.append(float(lines[-1].split()[-1][:-1]))
plt.plot(list(map(float, n_list)), t_list, marker="s", label="Binary")

n_list = ["027", "081", "243"]
t_list = []
std_list = []
for n in n_list:
    with open(f"vsN/ps1_ternary_{n}.log") as fin:
        lines = fin.readlines()
    t_list.append(float(lines[-1].split()[-2][1:-1]))
    std_list.append(float(lines[-1].split()[-1][:-1]))
plt.plot(list(map(float, n_list)), t_list, marker="d", label="Ternary")

n_list = ["004", "008", "016", "032", "064", "128", "256"]
plt.plot(list(map(float, n_list)), 0.05 * np.array(list(map(float, n_list))), linestyle="--", color="black")
n_list = ["027", "081", "243"]
plt.plot(list(map(float, n_list)), 2*np.array(list(map(float, n_list))), linestyle="--", color="black")
plt.text(40, 1.2, r"$\propto N_{\rm{mode}}$")
plt.text(50, 70, r"$\propto N_{\rm{mode}}$")


plt.title("(a)", loc="left")
plt.legend(fontsize="medium")
plt.xscale("log")
plt.yscale("log")
plt.xlabel("Number of modes $N_b$")
plt.ylabel("Time (s)")

#########################################################
plt.sca(axis[1])

m_list = ["005", "010", "020", "030", "050", "070", "100", "200", "300"]
m_array = np.array(list(map(float, m_list)))
t_list = []
std_list = []
for m in m_list:
    with open(f"vsM/ps1_linear_{m}.log") as fin:
        lines = fin.readlines()
    t_list.append(float(lines[-1].split()[-2][1:-1]))
    std_list.append(float(lines[-1].split()[-1][:-1]))
plt.plot(m_array, t_list, marker="o", label="MPS")


m_list = ["010", "020", "030", "050", "070", "100"]
m_array = np.array(list(map(float, m_list)))
t_list = []
std_list = []
for m in m_list:
    with open(f"vsM/ps1_binary_{m}.log") as fin:
        lines = fin.readlines()
    t_list.append(float(lines[-1].split()[-2][1:-1]))
    std_list.append(float(lines[-1].split()[-1][:-1]))
plt.plot(m_array, t_list, marker="s", label="Binary")

m_list = ["010", "020", "030", "040", "050", "060"]
m_array = np.array(list(map(float, m_list)))
t_list = []
std_list = []
for m in m_list:
    with open(f"vsM/ps1_ternary_{m}.log") as fin:
        lines = fin.readlines()
    t_list.append(float(lines[-1].split()[-2][1:-1]))
    std_list.append(float(lines[-1].split()[-1][:-1]))
plt.plot(m_array, t_list, marker="d", label="Ternary")

ref_array = np.array([50, 70, 100, 200, 300, 350])
plt.plot(ref_array, 1e-5 * ref_array**3, linestyle="--", color="black")
plt.text(200, 30, r"$\propto M_{\textrm{S}}^3$")
ref_array = np.array([50, 70, 100, 140])
plt.plot(ref_array, 2e-6 * ref_array**4, linestyle="--", color="black")
plt.text(120, 1000, r"$\propto M_{\textrm{S}}^4$")
ref_array = np.array([40, 70])
plt.plot(ref_array, 3e-6 * ref_array**5, linestyle="--", color="black")
plt.text(70, 3000, r"$\propto M_{\textrm{S}}^5$")

plt.title("(b)", loc="left")
plt.xscale("log")
plt.yscale("log")
plt.xlabel(r"Bond dimension $M_\textrm{S}$")
plt.ylabel("Time (s)")
plt.legend(fontsize="medium")


######################################################
plt.sca(axis[2])
d_list = ["005", "010", "020", "030", "050", "070", "100"]
#d_list = np.array(list(map(float, d_list)))
t_list = []
std_list = []
for d in d_list:
    with open(f"vsd/ps1_linear_{d}.log") as fin:
        lines = fin.readlines()
    t_list.append(float(lines[-1].split()[-2][1:-1]))
    std_list.append(float(lines[-1].split()[-1][:-1]))
plt.plot(np.array(list(map(float, d_list))), t_list, marker="o", label="MPS")

t_list = []
std_list = []
for d in d_list:
    with open(f"vsd/ps1_binary_{d}.log") as fin:
        lines = fin.readlines()
    t_list.append(float(lines[-1].split()[-2][1:-1]))
    std_list.append(float(lines[-1].split()[-1][:-1]))
plt.plot(np.array(list(map(float, d_list))), t_list, marker="s", label="Binary")

t_list = []
std_list = []
d_list = ["005", "010", "021", "030", "050", "070", "100"]
for d in d_list:
    with open(f"vsd/ps1_ternary_{d}.log") as fin:
        lines = fin.readlines()
    t_list.append(float(lines[-1].split()[-2][1:-1]))
    std_list.append(float(lines[-1].split()[-1][:-1]))
plt.plot(np.array(list(map(float, d_list))), t_list, marker="d", label="Ternary")

ref_array = np.array([50, 100])
plt.plot(ref_array, 2.5e-3 * ref_array**2, linestyle="--", color="black")
plt.text(60, 20, r"$\propto d^2$")

ref_array = np.array([30, 100])
plt.plot(ref_array, np.ones(len(ref_array))*2.5, linestyle="--", color="black")
plt.text(50, 1.7, r"$\propto \textrm{const}$")

ref_array = np.array([20, 100])
plt.plot(ref_array, np.ones(len(ref_array))*100, linestyle="--", color="black")
plt.text(50, 70, r"$\propto \textrm{const}$")


plt.title("(c)", loc="left")
plt.xscale("log")
plt.yscale("log")

plt.xlabel(r"Primitive basis $d$")
plt.ylabel("Time (s)")
plt.legend(fontsize="medium", loc=(0.3, 0.5))

plt.tight_layout()
plt.savefig("scaling.pdf")
```

```python
x_str_list = ["004", "008", "016", "032", "064", "128", "256", "512", "1024", "2048", "3072", "4096", "5120", "6144", "7068", "8092"]
x_list = list(map(float, x_str_list))
time_list = []
for x_str in x_str_list:
    with open(f"vsN/ps1_binary_{x_str}.log") as fin:
        text = fin.readlines()
    time = float(text[-1].split()[-2][1:-1])
    time_list.append(time)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression(fit_intercept=False)
lr.fit(np.array(x_list).reshape(-1, 1), time_list)
```

```python
lr.coef_
```

```python
plt.plot(x_list, time_list, marker="o", linestyle="", markerfacecolor="white", label="TTNS Simulation")
plt.plot(x_list, lr.coef_*np.array(x_list), label="Linear Fitting")
plt.legend()
plt.xlabel("Number of modes $N_b$")
plt.ylabel("Wall Time (s)")
plt.tight_layout()
plt.savefig("wall_time.pdf")
```
